```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits
import glob

from photutils.centroids import centroid_com, centroid_1dg, centroid_2dg
from image_registration import chi2_shift
from image_registration.fft_tools import shift
```

```python
def median_frames(folder, filenames, shape=None):

    ## Open test file to figure out shape
    if shape == None:
        hdul = fits.open(f'{folder}/{filenames[0]}.fits')
        shape = np.shape(hdul[0].data)

    ## Build [px*px*nframes] cube of frames
    cube = np.zeros((len(filenames), shape[0], shape[1]))

    for i, f in enumerate(filenames):

        hdul = fits.open(f'{folder}/{f}.fits')
        cube[i, :, :] = hdul[0].data
        hdul.close()

    ## Median cube along nframes axis
    cube_med = np.median(cube, axis=0)

    return cube_med
```

```python
folder = '/Users/isabelkain/Downloads/AO_data'
filelist = glob.glob(f'{folder}/*.fits')

## Generate list of filenames

fnames = np.array([])
fpaths = np.array([])

for f in filelist:
    txt = f.split('/')[-1].split('.')[0]  # reduce full path to just file name (w/out extension)
    fnames = np.append(fnames, txt)
    fpaths = np.append(fpaths, f[:-11])

## List of filetypes (e.g. bias, flat, science, etc)

ftypes = np.array([])

for f in filelist:

    hdul = fits.open(f)
    tag = hdul[0].header['OBJECT']
    hdul.close()

    ftypes = np.append(ftypes, tag)

## Zip filenames and types into pandas DF

filedict = pd.DataFrame(np.array([fpaths, fnames, ftypes]).T, columns=['Path', 'Filename', 'Target'])
filedict.sort_values(by='Filename', inplace=True, ignore_index=True)


## Manually include filter info

for i in range(10,17):  
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_JCH4'
    
for i in range(17, 25):
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_J'
    
for i in range(25, 32):
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_H'
    
for i in range(32, 37):
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_J'
    
for i in range(37, 42):
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_JCH4'
    
for i in range(42, 47):
    t = filedict.loc[i, 'Target']
    filedict.loc[i, 'Target'] = t + '_H'
    
filedict
```

```python
dark = median_frames(folder, filedict.loc[filedict['Target']=='dark', 'Filename'].values) / 45.
skyflat_J = median_frames(folder, filedict.loc[filedict['Target']=='sky_flat_J', 'Filename'].values) / 5.
skyflat_JCH4 = median_frames(folder, filedict.loc[filedict['Target']=='sky_flat_JCH4', 'Filename'].values) / 15.
skyflat_H = median_frames(folder, filedict.loc[filedict['Target']=='sky_flat_H', 'Filename'].values) / 7.5
```

```python
## Build counts/second science cube for each filter

nJ = np.sum(filedict['Target']=='Titan_J')
nJCH4 = np.sum(filedict['Target']=='Titan_J')
nH = np.sum(filedict['Target']=='Titan_J')


scicube_J = np.zeros((nJ, 2048, 2048))
scicube_JCH4 = np.zeros((nJCH4, 2048, 2048))
scicube_H = np.zeros((nH, 2048, 2048))


for i, f in enumerate(filedict.loc[filedict['Target']=='Titan_J', 'Filename']):
    
    hdul = fits.open(f'{folder}/{f}.fits')
    scicube_J[i, :, :] = hdul[0].data / 30.
    hdul.close()    


for i, f in enumerate(filedict.loc[filedict['Target']=='Titan_JCH4', 'Filename']):
    
    hdul = fits.open(f'{folder}/{f}.fits')
    scicube_JCH4[i, :, :] = hdul[0].data / 30.
    hdul.close()


for i, f in enumerate(filedict.loc[filedict['Target']=='Titan_H', 'Filename']):
    
    hdul = fits.open(f'{folder}/{f}.fits')
    scicube_H[i, :, :] = hdul[0].data / 20.
    hdul.close()
```

```python
## Calculate background

bkg_J = np.median(scicube_J, axis=0)
bkg_JCH4 = np.median(scicube_JCH4, axis=0)
bkg_H = np.median(scicube_H, axis=0)
```

```python
## Remove background from each image

scicube_J = scicube_J #- bkg_J # subtracting background leads to weird dark speckles
scicube_JCH4 = scicube_JCH4 #- bkg_JCH4
scicube_H = scicube_H #- bkg_H

fig, ax = plt.subplots(1,3, figsize=(8,4))

ax[0].imshow(scicube_J[0], vmin=0)
ax[1].imshow(scicube_JCH4[0], vmin=0)
ax[2].imshow(scicube_H[0], vmin=0)

ax[0].set_xlim(800, 1300)
ax[0].set_ylim(500, 1000)
ax[1].set_xlim(800, 1300)
ax[1].set_ylim(500, 1000)
ax[2].set_xlim(800, 1300)
ax[2].set_ylim(500, 1000)
```

## J-band alignment

```python
## Subtract two dithered images

sub1 = scicube_J[0] - scicube_J[1]
sub2 = scicube_J[1] - scicube_J[2]
sub3 = scicube_J[2] - scicube_J[3]
sub4 = scicube_J[3] - scicube_J[4]
sub5 = scicube_J[4] - scicube_J[0]

fig, ax = plt.subplots(1,5,figsize=(16,5))

ax[0].imshow(sub1)
ax[1].imshow(sub2)
ax[2].imshow(sub3)
ax[3].imshow(sub4)
ax[4].imshow(sub5)

## Set negative pixels equal to zero (to help centroid_2dg)

sub1[sub1 < 0.] = 0.
sub2[sub2 < 0.] = 0.
sub3[sub3 < 0.] = 0.
sub4[sub4 < 0.] = 0.
sub5[sub5 < 0.] = 0.

## Find centroid of each subtracted image

# xm1, ym1 = centroid_2dg(sub1**2)
# xm2, ym2 = centroid_2dg(sub2**2)
# xm3, ym3 = centroid_2dg(sub3**2)
# xm4, ym4 = centroid_2dg(sub4**2)
# xm5, ym5 = centroid_2dg(sub5**2)

xm1 = np.argmax(np.sum(sub1**2, axis=0))
ym1 = np.argmax(np.sum(sub1**2, axis=1))

xm2 = np.argmax(np.sum(sub2**2, axis=0))
ym2 = np.argmax(np.sum(sub2**2, axis=1))

xm3 = np.argmax(np.sum(sub3**2, axis=0))
ym3 = np.argmax(np.sum(sub3**2, axis=1))

xm4 = np.argmax(np.sum(sub4**2, axis=0))
ym4 = np.argmax(np.sum(sub4**2, axis=1))

xm5 = np.argmax(np.sum(sub5**2, axis=0))
ym5 = np.argmax(np.sum(sub5**2, axis=1))

## Plot subtracted plots with centroids to check

fig, ax = plt.subplots(1,5,figsize=(16,5))

ax[0].imshow(sub1**2)
ax[1].imshow(sub2**2)
ax[2].imshow(sub3**2)
ax[3].imshow(sub4**2)
ax[4].imshow(sub5**2)

ax[0].scatter(xm1, ym1, marker='+')
ax[1].scatter(xm2, ym2, marker='+')
ax[2].scatter(xm3, ym3, marker='+')
ax[3].scatter(xm4, ym4, marker='+')
ax[4].scatter(xm5, ym5, marker='+')


## Cut 100x100 pixel postage stamp of each subtracted image around positive image of target

post1 = sub1[int(ym1-150):int(ym1+150), int(xm1-150):int(xm1+150)]
post2 = sub2[int(ym2-150):int(ym2+150), int(xm2-150):int(xm2+150)]
post3 = sub3[int(ym3-150):int(ym3+150), int(xm3-150):int(xm3+150)]
post4 = sub4[int(ym4-150):int(ym4+150), int(xm4-150):int(xm4+150)]
post5 = sub5[int(ym5-150):int(ym5+150), int(xm5-150):int(xm5+150)]

## Take centroid again to be precise
pcenx1, pceny1 = centroid_2dg(post1)
pcenx2, pceny2 = centroid_2dg(post2)
pcenx3, pceny3 = centroid_2dg(post3)
pcenx4, pceny4 = centroid_2dg(post4)
pcenx5, pceny5 = centroid_2dg(post5)

## Plot postage stamps to check clipping

fig, ax = plt.subplots(1,5,figsize=(16,5))

ax[0].imshow(post1)
ax[1].imshow(post2)
ax[2].imshow(post3)
ax[3].imshow(post4)
ax[4].imshow(post5)

ax[0].scatter(pcenx1, pceny1, marker='+')
ax[1].scatter(pcenx2, pceny2, marker='+')
ax[2].scatter(pcenx3, pceny3, marker='+')
ax[3].scatter(pcenx4, pceny4, marker='+')
ax[4].scatter(pcenx5, pceny5, marker='+')


## IF all good, then stack!

aligned_J = np.sum([post1, post2, post3, post4, post5], axis=0)

plt.figure()
plt.imshow(aligned_J)
plt.colorbar()
plt.title('Aligned J-band image of Titan')
```
