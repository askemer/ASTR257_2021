```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

#  Averaging Bias Frames

```python
directory = 'calib_data/'

hdul = fits.open(directory+'d1001.fits')
data_bias_1=hdul[0].data

hdul = fits.open(directory+'d1002.fits')
data_bias_2=hdul[0].data

hdul = fits.open(directory+'d1003.fits')
data_bias_3=hdul[0].data

hdul = fits.open(directory+'d1004.fits')
data_bias_4=hdul[0].data

hdul = fits.open(directory+'d1005.fits')
data_bias_5=hdul[0].data
```

```python
data_bias_1 = data_bias_1.astype(float)
data_bias_2 = data_bias_1.astype(float)
data_bias_3 = data_bias_1.astype(float)
data_bias_4 = data_bias_1.astype(float)
data_bias_5 = data_bias_1.astype(float)
```

```python
print(np.max(data_bias_1))
```

```python
bias_median = np.zeros_like(data_bias_1)
for i in range(1024):
    for j in range(1056):
        temp = np.array([data_bias_1[i,j],data_bias_2[i,j],data_bias_3[i,j],data_bias_4[i,j],data_bias_5[i,j]])

        bias_median[i,j] = np.median(temp)
```

```python
plt.imshow(bias_median)
```

#  Darks

```python
hdul = fits.open(directory+'d1016.fits') # 20 secs
data_dark_1=hdul[0].data

hdul = fits.open(directory+'d1017.fits') # 20 secs
data_dark_2=hdul[0].data

hdul = fits.open(directory+'d1018.fits') # 20 secs
data_dark_3=hdul[0].data

hdul = fits.open(directory+'d1019.fits') # 20 secs
data_dark_4=hdul[0].data

hdul = fits.open(directory+'d1020.fits') # 20 secs
data_dark_5=hdul[0].data


data_dark_1= (data_dark_1-bias_median)/20.0
data_dark_2= (data_dark_2-bias_median)/20.0
data_dark_3= (data_dark_3-bias_median)/20.0
data_dark_4= (data_dark_4-bias_median)/20.0
data_dark_5= (data_dark_5-bias_median)/20.0

dark_median = np.zeros_like(data_bias_1)
for i in range(1024):
    for j in range(1056):
        temp = np.array([data_dark_1[i,j],data_dark_2[i,j],data_dark_3[i,j],data_dark_4[i,j],data_dark_5[i,j]])

        dark_median[i,j] = np.median(temp)
```

# Averaging Sky Flats 

```python
hdul = fits.open(directory+'d1134.fits') # 5 secs
data_sflat_1=hdul[0].data

hdul = fits.open(directory+'d1135.fits') # 6 secs
data_sflat_2=hdul[0].data

hdul = fits.open(directory+'d1136.fits') # 10 secs
data_sflat_3=hdul[0].data

hdul = fits.open(directory+'d1137.fits') # 10 secs
data_sflat_4=hdul[0].data

hdul = fits.open(directory+'d1138.fits') # 10 secs
data_sflat_5=hdul[0].data

# normalize flats to 1 sec

data_sflat_1= (data_sflat_1-bias_median)/5.0
data_sflat_2= (data_sflat_2-bias_median)/6.0
data_sflat_3= (data_sflat_3-bias_median)/10.0
data_sflat_4= (data_sflat_4-bias_median)/10.0
data_sflat_5= (data_sflat_5-bias_median)/10.0

data_sflat_1 = data_sflat_1 - dark_median
data_sflat_2 = data_sflat_2 - dark_median
data_sflat_3 = data_sflat_3 - dark_median
data_sflat_4 = data_sflat_4 - dark_median
data_sflat_5 = data_sflat_5 - dark_median


data_sflat_1= data_sflat_1/np.max(data_sflat_1)
data_sflat_2= data_sflat_2/np.max(data_sflat_2)
data_sflat_3= data_sflat_3/np.max(data_sflat_3)
data_sflat_4= data_sflat_4/np.max(data_sflat_4)
data_sflat_5= data_sflat_5/np.max(data_sflat_5)

```

```python
flat_median = np.zeros_like(data_bias_1)
for i in range(1024):
    for j in range(1056):
        temp = np.array([data_sflat_1[i,j],data_sflat_2[i,j],data_sflat_3[i,j],data_sflat_4[i,j],data_sflat_5[i,j]])

        flat_median[i,j] = np.median(temp)
```
