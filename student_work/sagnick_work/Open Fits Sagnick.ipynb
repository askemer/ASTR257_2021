```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from astropy.nddata import Cutout2D
from astropy import units as u
```

```python


hdul = fits.open('test.fits')
```

```python
print(hdul.info())
```

```python
data  = hdul[0].data
dataplot = np.nan_to_num(data)
```

```python
plt.figure(figsize=(10,10))
im=plt.imshow(dataplot,norm=LogNorm(vmin=1e-1, vmax=np.max(dataplot)),cmap='Greys_r')
plt.colorbar(im)
```

```python
position = (55, 50)
size = (40, 40)     # pixels
cutout = Cutout2D(data, position, size)
```

```python
plt.figure(figsize=(10,10))
im=plt.imshow(cutout.data,norm=LogNorm(vmin=1e-1, vmax=np.max(cutout.data)),cmap='Greys_r')
plt.colorbar(im)
```

```python
plt.imshow(dataplot,norm=LogNorm(vmin=1e-1, vmax=np.max(cutout.data)),cmap='Greys_r')
cutout.plot_on_original(color='red')
```

```python
from astropy.stats import sigma_clipped_stats
from photutils.detection import DAOStarFinder

mean, median, std = sigma_clipped_stats(dataplot, sigma=3.0) 
print((mean, median, std))  
```

```python
daofind = DAOStarFinder(fwhm=3.0, threshold=5.*std)  
sources = daofind(dataplot - median)  
```

```python
for col in sources.colnames:  
     sources[col].info.format = '%.8g'  # for consistent table output
```

```python
print(sources)
```

```python
print(np.max(sources['peak']))
```

```python
wh = np.where(sources['peak']==np.max(sources['peak']))
```

```python
print(sources['xcentroid'][wh],sources['ycentroid'][wh])
```
